# üìä An√°lisis General de Resultados (15 min por instancia)
Este notebook analiza los resultados obtenidos con CPLEX en 15 minutos por instancia. Se utiliza el archivo `Comparacion_Resultados_15min.xlsx` como fuente principal.

```python
# üì¶ Cargar librer√≠as
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(style='whitegrid')
```

```python
# üìÅ Cargar archivo de comparaci√≥n
ruta_archivo = os.path.join("..", "notebooks", "Comparacion_Resultados_15min.xlsx")

df = pd.read_excel(ruta_archivo)
df
```

```python
# üìâ Gr√°fico: GAP por instancia
plt.figure(figsize=(10, 5))
sns.barplot(x='Instancia', y='GAP (%)', data=df, palette='coolwarm')
plt.title('GAP (%) por instancia (15 min)', fontsize=14)
plt.ylabel('GAP (%)')
plt.xlabel('Instancia')
plt.ylim(0, max(df['GAP (%)'].dropna()) + 10)
plt.tight_layout()
plt.show()
```

```python
# üìà Comparaci√≥n de objetivos alcanzados
plt.figure(figsize=(10, 5))
sns.barplot(x='Instancia', y='Objetivo', data=df, palette='crest')
plt.title('Objetivo alcanzado por instancia')
plt.ylabel('Valor objetivo')
plt.xlabel('Instancia')
plt.tight_layout()
plt.show()
```
