# üìä An√°lisis General de Resultados (15 min por instancia)
Este notebook analiza los resultados obtenidos con CPLEX en 15 minutos por instancia. Se utiliza el archivo `Comparacion_Resultados_15min.xlsx` como fuente principal.

```python
# üì¶ Cargar librer√≠as
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(style='whitegrid')
```

```python
# üìÅ Cargar archivo de comparaci√≥n
ruta_archivo = os.path.join("..", "notebooks", "Comparacion_Resultados_15min.xlsx")

df = pd.read_excel(ruta_archivo)
df
```

```python
# üìâ Gr√°fico: GAP por instancia
plt.figure(figsize=(10, 5))
sns.barplot(x='Instancia', y='GAP (%)', data=df, palette='coolwarm')
plt.title('GAP (%) por instancia (15 min)', fontsize=14)
plt.ylabel('GAP (%)')
plt.xlabel('Instancia')
plt.ylim(0, max(df['GAP (%)'].dropna()) + 10)
plt.tight_layout()
plt.show()
```

```python
# üìà Comparaci√≥n de objetivos alcanzados
plt.figure(figsize=(10, 5))
sns.barplot(x='Instancia', y='Objetivo', data=df, palette='crest')
plt.title('Objetivo alcanzado por instancia')
plt.ylabel('Valor objetivo')
plt.xlabel('Instancia')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

datos = [
    (1, 12.49, 0.00),
    (2, 71.28, 0.00),
    (3, 900.22, 0.1201),
    (4, 900.28, 0.0681),
    (5, 900, 0.3637),
    (6, 900.03, 0.2976),
    (7, 900.09, 0.8403),
    (8, 900.08, 0.5639),
    (10, 900, 0.5993)
]

# Extraer IDs, tiempos y GAPs
ids = [d[0] for d in datos]
times = [d[1] for d in datos]
gaps = [d[2] for d in datos]

plt.figure(figsize=(10, 5))
plt.plot(ids, times, 'o-', color="red", label="Tiempo por caso")
plt.fill_between(ids, times, alpha=0.2, color="orange")  # √Årea bajo la curva
plt.xlabel("N√∫mero de Instances")
plt.ylabel("Tiempo de soluci√≥n (s)")
plt.title("Tendencia de tiempo de ejecuci√≥n (mayor Instancia ‚Üí mayor tiempo)")
plt.xticks(ids)  # Mostrar todos los IDs en el eje X
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
